# 2차 데이터프레임 생성(파생변수 생성)

-파생변수를 위해 생성한 변수를 데이터프레임에서 제거

```python
def year_devide(x):

  x['SalesYear']=x['SalesYear'].astype(int)
  df_20 = x[x['SalesYear']==2020]
  df_21 = x[x['SalesYear']==2021]
  
  return df_20, df_21
```

# 날씨 관련 파생 변수 생성

```python
#비의 유무 변수 RAIN_NORAIN 생성
df_raw["RAIN_NORAIN"] = df_raw.apply(lambda x : 1 if x['RainfallAmount'] >= 1 else 0, axis = 1) #1이상이면 비 옴
```

```python
#비의 강도 변수 RAIN_STRENGTH 생성 
#강수량 데이터 구성을 확인하고 강수량이 있었을 때만 보고 기준 세우기
cnt_check = df_raw.query("RainfallAmount >= 1")
cnt_check["RainfallAmount"].describe()
#강수량 범위가 1.0 ~ 103.1 인것을 확인할 수 있고 가상청 강수량 기준을 근거로 강수 강도를 약:1 ~ 3미만 / 중:3 ~ 15미만 / 강:15  ~ 30미만 / 매우 강함:30 ~ 103
```

```python
from numpy import NaN
def rain_strength(x):
    if x >= 30:              
      return 'STRONG'
    
    elif x >= 15:       
      return 'SEMI_STRONG'
    
    elif x >= 3:
      return 'MEDIUM'
      
    elif x >= 1:
      return 'WEAK'

    else:
      return NaN
       
df_raw['RAIN_STRENGTH'] = df_raw['RainfallAmount'].apply(rain_strength)
```

```python
#적설 여부 변수 SNOW_NOSNOW 생성
df_raw["SNOW_NOSNOW"] = df_raw.apply(lambda x: 1 if x["SnowloadMax"] >= 1 else 0, axis=1) #1이상이면 눈
```

```python
# 날씨 변수 WEATHER 생성
def weather(x, y):
  if x == 1 and y ==0: #비가 오고 눈이 안오면
        return 'RAIN'
  
  elif x ==0 and y ==1: #비가 안오고 눈이 오면
        return 'SNOW'
  
  elif x == 1 and y ==1: #비도 오고 눈도 오면
        return 'SNOW'
  
  else:
        return 'NORMAL'
df_rain_snow = df_raw[['RAIN_NORAIN','SNOW_NOSNOW']]
df_raw['WEATHER'] = df_rain_snow.apply(lambda x: weather(x.RAIN_NORAIN, x.SNOW_NOSNOW), axis = 1)
```

```python
#온도 범주화 변수 TEMP_CAT 생성

def temp_cat(x):
    if x >= 23.5:              
      return 'HOT'
    
    elif x >= 15:       
      return 'WARM'
    
    elif x >= 7.6:
      return 'COOL'
      
    else:
      return 'COLD'
       
df_raw['TEMP_CAT'] = df_raw['TempAvg'].apply(temp_cat)
```

# 고객 관련 파생 변수 생성

```python
#고객 충성도 변수 CUST_ROYALTY 생성
#idUser별 방문 횟수 변수 visit_cnt 생성
#2020
df_raw_2020 = df_raw.query("SalesYear == 2020")
df_raw_2020 = df_raw_2020.groupby(['idPOS','idUser'], as_index= False).count()
# df_raw_2020['idUser'].value_counts().describe() ##방문 횟수 확인
visit_count_2020 = pd.DataFrame({'visit_cnt': df_raw_2020['idUser'].value_counts()}) #방문 횟수 데이터프레임 저장
visit_count_2020.reset_index(inplace=True)
visit_count_2020 = visit_count_2020.rename(columns={'index':'idUser'})
visit_count_2020['SalesYear'] = 2020
visit_count_2020
```

```python
df_raw_2020 = pd.merge(df_raw, visit_count_2020, on=["idUser","SalesYear"], how='inner') #df_raw에서 2020년 부분만 가져옴
```

```python
#idUser별 방문 횟수 확인
#2021
df_raw_2021 = df_raw.query("SalesYear == 2021")
df_raw_2021 = df_raw_2021.groupby(['idPOS','idUser'], as_index= False).count()
# df_raw_2021['idUser'].value_counts().describe() #방문 횟수 확인
visit_count_2021 = pd.DataFrame({'visit_cnt': df_raw_2021['idUser'].value_counts()}) #방문 횟수 데이터프레임 저장

visit_count_2021.reset_index(inplace=True)
visit_count_2021 = visit_count_2021.rename(columns={'index':'idUser'})
visit_count_2021['SalesYear'] = 2021
visit_count_2021
```

```python
df_raw_2021 = pd.merge(df_raw, visit_count_2021, on=["idUser","SalesYear"], how='inner')  #df_raw에서 2021년 부분만 가져옴
```

```python
df_raw = pd.merge(df_raw_2020, df_raw_2021, how='outer') #2021, 2022 병합
```

```python
df_raw["visit_cnt"].describe()
```

```python
def grade(x):
    if x >= 198:              #198번 이상 방문
        return 'A'
    
    elif x >= 100:           #100~198번 방문
        return 'B'
    
    else:                         #99번 이하
        return 'C'
    
df_raw['Grade'] = df_raw['visit_cnt'].apply(grade)
```

```python
# 일별 객단가 변수 DAY_CT 생성
ct = df_raw.groupby(['idPOS', 'SalesDate'])['Price'].sum().reset_index()
df_ct = pd.DataFrame(ct.groupby(['SalesDate'], as_index=False)['Price'].mean().round(0))
df_ct.rename(columns = {"Price":"DAY_CT"}, inplace=True)
df_raw = pd.merge(df_raw, df_ct, on='SalesDate', how='outer')
```

# 매출 관련 파생변수 생성

```python
# 일별 매출 변수 DAY_AMOUNT 생성 
df_day_amount = pd.DataFrame(df_raw.groupby(['SalesDate'])['Price'].sum())
df_day_amount.rename(columns = {'Price':'DAY_AMOUNT'}, inplace=True)
df_day_amount.reset_index()
df_raw = pd.merge(df_raw, df_day_amount, on='SalesDate', how='outer')
df_raw
```

```python
# 월별 매출액 MONTH_AMOUNT 생성
df_month_amount = pd.DataFrame(df_raw.groupby(['SalesYearMonth'])['Price'].sum())
df_month_amount.rename(columns = {'Price':'MONTH_AMOUNT'}, inplace=True)
df_month_amount.reset_index()
df_raw = pd.merge(df_raw, df_month_amount, on='SalesYearMonth', how='outer')
df_raw
```

```python
#요일별 매출액 변수 DOW_AMOUNT 생성
#요일을 수치형으로 표현하는 변수 SalesDow_num 생성
# 월 = 0, 화 = 1, 수 = 2... 일 = 6
df_raw['SalesDate'] = pd.to_datetime(df_raw['SalesDate'], format='%Y-%m-%d', errors='raise')
df_raw["SalesDow_num"] = df_raw["SalesDate"].dt.dayofweek
```

```python
df_raw['SalesYear'].dtype
```

```python
df_raw['SalesYear']=df_raw['SalesYear'].astype(str)
df_raw['SalesYearDow'] = df_raw['SalesYear'] + df_raw['SalesDow']
df_raw
```

```python
# 요일별 매출액
df_dow_amount = pd.DataFrame(df_raw.groupby(['SalesYearDow'])['Price'].sum())
df_dow_amount.rename(columns = {'Price':'DOW_AMOUNT'}, inplace=True)
df_dow_amount.reset_index()
df_raw = pd.merge(df_raw, df_dow_amount, on='SalesYearDow', how='outer')
df_raw
```

```python
# #2020/2021로 데이터셋 구분
df_20, df_21 = year_devide(df_raw)
df_20
```

```python
#연간 시간대별 매출액 변수인 HOUR_AMOUNT 생성
a = pd.DataFrame(df_raw.groupby(["SalesYear", "SalesHour"])['Price'].sum())
a.rename(columns={'Price':'HOUR_AMOUNT'}, inplace=True)
df_raw = pd.merge(df_raw, a, on=['SalesYear', 'SalesHour'])
```

```python
#연령대에서 '대'를 뺀 변수 AgeGroup_num생성
def remove_int(x):
  if '대이상' in x:
    fixed_value = x.replace('대이상','')
  else:
    fixed_value = x.replace('대','')
  fixed_value = int(fixed_value)
  return fixed_value

# df_raw["AgeGroup_num"] = df_raw.apply(lambda x: x["AgeGroup"] else 0, axis=1) #1이상이면 눈
  # print(fixed_value)


df_raw['AgeGroup_num'] = df_raw['AgeGroup'].apply(remove_int)
```

```python
df_raw_copy = df_raw.copy()
```

```python
#1-1. 30대/남/7+8/가공식사 + 음료
df_sub_4asso = df_raw_copy[(df_raw_copy["SalesYear"] == 2021) & (df_raw_copy["Gender"]=="남성") & (df_raw_copy["AgeGroup_num"]== 30) & 
                            ((df_raw_copy["ItemNameMiddle"] == "가공식사") | (df_raw_copy["ItemNameMiddle"] == "음료")) &
                            ((df_raw_copy["SalesHour"] == 7) | (df_raw_copy["SalesHour"] == 8))]

df_sub_4asso
ID = list(set(df_sub_4asso["idUser"]))

_list_asso = []
for i in ID:
    # ID별 구매 데이터 추출
    tmp_list = list(df_sub_4asso[df_sub_4asso["idUser"] == i]["ItemName"])
    tmp_list.sort()
    # ID별 구매 데이터를 리스트에 담기
    _list_asso.append(tmp_list)

enc = TransactionEncoder() # 시간 많이 걸려서
df_raw_enc = enc.fit_transform(X = _list_asso)
df_asso = pd.DataFrame(df_raw_enc, columns = enc.columns_)
```

```python
v_min_support = 0.013
df_freq = apriori(df_asso, min_support = v_min_support, use_colnames = True)
df_freq.round(3)
```
